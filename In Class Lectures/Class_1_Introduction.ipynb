### 🤖 **Example 1: A Sentiment Analyzer**

In our survey, one of the terms I asked about was 'Sentiment Analysis'. This is essentially an analysis of the 'sentiment' of something - of a sentence, of a social media post, of a book, of a picture, anything.



#### A "Rule-based" Sentiment Analyzer



Below is a short function that undertakes a simple sentiment analysis.

**Run the cell below and type a sentence when prompted.**

```python
from textblob import TextBlob

# This function calculates a score based on a pre-set dictionary
def check_sentiment():
    sentence = input("Type a sentence to test the simple AI: ")
    score = TextBlob(sentence).sentiment.polarity
    print(f"\nSimple AI Score: {score}")

    if score > 0:
        print("Result: POSITIVE")
    elif score < 0:
        print("Result: NEGATIVE")
    else:
        print("Result: NEUTRAL")

check_sentiment()
```

This simple code above is not "real AI". Why? Because it doesn't "think." It just has a dictionary of words with assigned scores (e.g., "Great" = +1, "Awful" = -1). It just does the math.

---

As comparison: Imagine a librarian who doesn't speak a word of English. To help people, they have two lists:
* **List A (Positive):** "Great", "Excellent", "Happy"
* **List B (Negative):** "Awful", "Bad", "Sad"

When you give them a book, they count the words. If there are more "List A" words, they put a smiley face on the cover. This is exactly how this script works. It is called a **Lexicon-based approach**.

#### A "Real" Sentiment Analyzer



```python
# 1. Install the "Transformers" library (The "App Store" for AI models)
!pip install -q transformers

# 2. Import the "Pipeline" (The "Shortcut" to use a pre-trained AI brain)
from transformers import pipeline

# 3. Load a specific "Brain" for understanding emotions
# Note: This may take a minute to download the first time!
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 4. Create an INPUT BOX (This lets the student type their own sentence)
user_sentence = input("Type a sentence to test the AI's feelings: ")

# 5. Ask the AI for its opinion on what the student typed
result = classifier(user_sentence)[0]

# 6. Show the results
print(f"\nYour Sentence: {user_sentence}")
print(f"AI Verdict: {result['label']}")
print(f"Confidence Level: {round(result['score'] * 100, 2)}%")
```

Unlike the dictionary version, this AI doesn't just look for "good" or "bad" words. It understands the **relationships** between words.

##### Key Differences:
* **Context over Content:** In a dictionary, "great" is always +1. In this AI, the meaning of "great" will depend on the context within which great was written. For instance, the AI will know that *"Great, another error"* is actually negative (try it!).
* **Probability, not Math:** This AI doesn't give a "score." It calculates **Probability**. When it says `POSITIVE`, it’s saying: *"Based on the millions of sentences I've read, there is a 99% chance this is a happy sentence."*
* **Transfer Learning:** This model (DistilBERT) was already "pre-trained" on the entirety of English Wikipedia and thousands of books. This pretraining is different than the dictionary. No human told the AI that "great" is good. Rather, the AI read these thousands of books and noticed that "great" usually appears near words like "Happy," "Success," and "Winning." It then "intuited" the value of great itself. This is called Self-Supervised Learning.

This is the difference between a Stochastic Parrot (repeating rules) and Machine Learning (understanding patterns). The AI "learned" patterns from data rather than following a list of rules written by a human.

### 🤖 **Example 2: A Text Generator**

We can also demonstrate the differences between a rule-based system and an AI system with regards to a simple text generation procedure.



#### A "Rule-based" Text Completer

This is really just a "Mad Libs" generator. It picks random words from a list you provide.

This is *Deterministic*. It can only say what you told it to say. It has zero "intelligence."

```python
import random
names = ["The AI", "A Robot", "A Human", "A Frog", "A Rock"]
verbs = ["eats", "thinks about", "learns", "carries", "hates"]
nouns = ["data", "pizza", "the world", "money", "volcanos"]
print(f"{random.choice(names)} {random.choice(verbs)} {random.choice(nouns)}.")
```

#### A "Real" Text Completer (Generative AI)

This uses GPT-2 (the ancestor of ChatGPT).

It is *Probabilistic*. It is actually "thinking" (predicting) the next word based on patterns it learned from its training set.

```python
from transformers import pipeline

# 2. Load the Brain (GPT-2)
# Note: This might take a minute the first time!
generator = pipeline('text-generation', model='gpt2')

# 3. Get student input
prompt = input("Start a story (e.g., 'In the future, robots will...'): ")

# 4. Generate the story
result = generator(prompt, max_new_tokens=100, truncation=True, no_repeat_ngram_size=2, pad_token_id=50256)

# 5. Show ONLY the story
print("\n--- AI'S FINISHED STORY ---")
print(result[0]['generated_text'])
print("---------------------------")
```
